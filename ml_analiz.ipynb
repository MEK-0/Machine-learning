Yapay zeka tabanlı analiz sonuçlarına dayalı olarak oluşturulan tahmini grafikler, sistemin devreye alınmasıyla sağlanan iyileştirmeleri çarpıcı bir şekilde ortaya koymaktadır. Grafikler; hata tespit başarısı, bakım maliyetleri ve üretim duruş süreleri gibi kritik metriklerdeki değişimleri göstermektedir. Sistem olmadan önce %60 olan hata tespit oranının yapay zeka destekli erken tespit algoritmaları sayesinde %85’e yükseldiği gözlenmiştir. Aynı zamanda, yıllık bakım maliyetleri %20 azalarak 100.000 TL’den 80.000 TL’ye, üretim duruş süreleri ise %40 azalarak 50 saatten 30 saate düşmüştür. Bu grafikler, yapay zekanın üretim süreçlerinde hem maliyet hem de zaman açısından sağladığı önemli katkıları görselleştirmekte ve projenin etkinliğini öne çıkarmaktadır.

```python
# Gerekli kütüphaneleri yükleyelim
import matplotlib.pyplot as plt

# Örnek veri setini tanımlayalım
veriler = {
    "Metrik": ["Başarı", "Maliyet", "Zaman"],
    "Sistem Öncesi": [60, 100000, 50],
    "Sistem Sonrası": [85, 80000, 30]
}

# Grafik boyutunu ayarlayalım
plt.figure(figsize=(12, 5))

# Her bir metriği karşılaştıran çubuk grafikleri oluşturalım
for i, metrik in enumerate(veriler["Metrik"]):
    plt.subplot(1, 3, i + 1)  # Her metriğe ayrı bir grafik
    plt.bar(["Sistem Öncesi", "Sistem Sonrası"], [veriler["Sistem Öncesi"][i], veriler["Sistem Sonrası"][i]], color=['red', 'green'])
    plt.title(metrik)
    if metrik == "Başarı":
        plt.ylabel("Oran (%)")
    elif metrik == "Maliyet":
        plt.ylabel("Maliyet (TL)")
    elif metrik == "Zaman":
        plt.ylabel("Duruş Süresi (Saat)")

plt.tight_layout()
plt.show()
```
