#basit regresyon ile otomobil fiyat tahmini


#kütüphaneler



```python
import numpy as np
import matplotlib.pyplot as plt

```

#veri seti hazırlama


```python
x=np.array([[6000],[8200],[9000],[14200],[16200]]).reshape(-1,1)
y=[86000,82000,78000,75000,70000]
```

#öğrenme veri setini görselleştirme


```python
plt.figure()
plt.title('otomobil fiyat-KM dağılım grafiği')
plt.xlabel('KM')
plt.ylabel('fiyat')
plt.plot(x,y,'k.')
plt.axis([30000,20000,60000,95000])
plt.grid(True)
plt.show()
```

#regresyon tahmini için kütüphane ekleme


```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

#algoritma model parametreleri

```python
model.intercept_
model.coef_
```

#modeli tahmin doğrusu

```python
plt.figure()
plt.title('otomobil fiyat-KM dağılım grafiği')
plt.xlabel('KM')
plt.ylabel('fiyat')
plt.plot(x,model.predict(x),color = 'red')
plt.plot(x,y,' k.')
plt.grid(True)
plt.show()
```

#12.000 km lik aracın fiyat tahmini

```python
test_araba=np.array([[12000]])
predicted_price=model.predict(test_araba)[0]
print("12000 km lik arabanın tahmini fiyatı :",predicted_price," tldir")
```

#9000 km lik aracın fiyat tahmini

```python
test_Araba2=np.array([[9000]])
predicted_price2=model.predict(test_Araba2)[0]
print("9000 km lik aracın tahmini fiyatı",predicted_price2," tldir")
```

# y-prediction isimli bir değişken oluşturup araçların km'lerinden hareketle fiyat karşılaştırma


```python
y_prediction = model.predict(x)
for i, prediction in enumerate(y_prediction):
   print("tahmin edilen fiyat : %2f , gerçek fiyat: %s"%(prediction,y[i]))
```
