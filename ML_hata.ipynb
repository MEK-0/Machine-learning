
# Kütüphane kurulumu




```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Sayısal Veri Tablosunun Görselleştirilmesi

```python
data = {
    'Gösterge': [
        'Ortalama Yıllık Hata Oranı (%)',
        'Hata Tespit Süresi (saatten saniyeye)',
        'Yıllık Bakım Maliyeti (USD)',
        'Üretim Duruş Oranı (%)'
    ],
    'Mevcut Durum': [15, 24, 50000, 10],
    'Proje Sonrası Hedef': [5, 0.0014, 40000, 6],
    'İyileşme Oranı (%)': [66, 99, 20, 40]
}

# DataFrame
df = pd.DataFrame(data)

# Çubuk grafiği
fig, ax = plt.subplots(figsize=(12, 7))
width = 0.35  # Çubuk genişliği
x = np.arange(len(df['Gösterge']))

# Mevcut durum ve proje sonrası değerler
ax.bar(x - width/2, df['Mevcut Durum'], width, label='Mevcut Durum', color='skyblue')
ax.bar(x + width/2, df['Proje Sonrası Hedef'], width, label='Proje Sonrası Hedef', color='salmon')

# Grafik düzenlemeleri
ax.set_xlabel('Gösterge')
ax.set_ylabel('Değer')
ax.set_title('Mevcut Durum ve Proje Sonrası Hedefler')
ax.set_xticks(x)
ax.set_xticklabels(df['Gösterge'], rotation=45, ha="right")
ax.legend()

# İyileşme oranını göstermek için ikinci bir y ekseni
ax2 = ax.twinx()
ax2.plot(x, df['İyileşme Oranı (%)'], color='green', marker='o', linestyle='--', label='İyileşme Oranı (%)')
ax2.set_ylabel('İyileşme Oranı (%)', color='green')

plt.tight_layout()
plt.show()
```

#Hata Tespit Doğruluk Oranı Grafiği

```python
# Yıllara göre doğruluk oranı verisi
years = np.array([2018, 2019, 2020, 2021, 2022])
accuracy = np.array([70, 75, 80, 90, 95])  # Yüzdelerle doğruluk oranları

# Çizim
plt.figure(figsize=(8, 6))
plt.plot(years, accuracy, marker='o', color='b', linestyle='-', linewidth=2, markersize=6)
plt.xlabel('Yıl')
plt.ylabel('Doğruluk Oranı (%)')
plt.title('Tahmine Dayalı Bakım Doğruluk Oranı')
plt.grid(True)
plt.show()
```

# Hedeflenen İyileştirme Oranı Pasta Grafiği

```python
# Veriler
labels = ['Ortalama Yıllık Hata Oranı İyileşmesi', 'Hata Tespit Süresi İyileşmesi',
          'Yıllık Bakım Maliyeti Azaltılması', 'Üretim Duruş Oranı Azaltılması']
sizes = [66, 99, 20, 40]  # İyileşme oranları
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

# Pasta grafiği çizimi
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Hedeflenen İyileştirme Oranları')
plt.show()
```
